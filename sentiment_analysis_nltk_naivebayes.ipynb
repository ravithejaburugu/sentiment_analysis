```python
import nltk
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import subjectivity
from nltk.sentiment import SentimentAnalyzer
from nltk.sentiment.util import *
```

```python
n_instances = 100
subj_docs = [(senti, 'subj') for senti in subjectivity.sents(categories='subj')[:n_instances]]
subj_docs
```

```python
obj_docs = [(senti, 'obj') for senti in subjectivity.sents(categories='obj')[:n_instances]]
obj_docs
```

```python
print(len(subj_docs), len(obj_docs))
print(subj_docs[0])
```

```python
train_subj_docs = subj_docs[:80]
test_subj_docs = subj_docs[80:]
train_obj_docs = obj_docs[:80]
test_obj_docs = obj_docs[80:]
```

```python
train_docs = train_subj_docs + train_obj_docs
test_docs = test_subj_docs + test_obj_docs
```

```python
senti_analyzer = SentimentAnalyzer()
```

```python
all_words_neg = senti_analyzer.all_words([mark_negation(doc) for doc in train_docs])
all_words_pos = senti_analyzer.all_words([mark_positive(doc) for doc in train_docs])
```

```python
unigram_feats = senti_analyzer.unigram_word_feats(all_words_neg, min_freq=4)
```

```python
senti_analyzer.add_feat_extractor(extract_unigram_feats, unigrams=unigram_feats)
```

```python
train_set = senti_analyzer.apply_features(train_docs)
test_set = senti_analyzer.apply_features(test_docs)
```

```python
trainer = NaiveBayesClassifier.train
```

```python
classifier = SentimentAnalyzer.train(trainer=trainer, training_set=train_set)
```

```python
for key, value in sorted(senti_analyzer, evaluate(test_set).items()):
    print('{0}:{1}'.format(key, value))
```
